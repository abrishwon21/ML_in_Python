```
import matplotlib.pyplot as plt
import numpy as np
fig,ax=plt.subplots(1,1)
marks=np.array([20,60,25,30,19,35,38,45,50,60,55,51,81,89,70,78,90,95,61,66,53])
range=[0,20,40,60,80,100]
#ax.hist(marks,bins=range,histtype='step')
#or
ax.hist(marks,bins=range,color=['c'])
ax.set_title('histogram')
ax.set_xlabel('bins')
ax.set_xticks([20,40,80,95])
ax.set_yticks([1,3,5,7])
ax.set_ylabel('no of students')
plt.show()
```

```
nofuusers=[100,200,90,70,80]
pl=['java','python','c','c++','javascript']
plt.axis('equal')
plt.pie(nofuusers,labels=pl,autopct='%.2f%%',shadow=True)
plt.show()#to show or extract the specific parts
```

```
#explode concept and legend
nofuusers=[100,200,90,70,80]
pl=['java','python','c','c++','javascript']
plt.axis('equal')
plt.pie(nofuusers,labels=pl,autopct='%.2f%%',shadow=True,explode=[0.08,0.1,0,0,0])
plt.legend(loc='lower left',ncol=5)
plt.show()#to show or extract the specific parts
```

```
c1=np.random.normal(100,10,200)
c2=np.random.normal(100,10,200)
c3=np.random.normal(70,10,200)
c4=np.random.normal(90,20,200)
fig=plt.figure()
c=['g','r','b','r']
ax=fig.add_axes([0,0,1,1])
ax.boxplot([c1,c2,c3,c4])
plt.legend(loc='upper left')
plt.show()
```

```
boys=[80,56,78,90,75,45,55,67,80,55,52,88,92]
girls=[99,89,94,84,56,78,90,66,77,88,92,80,61]
fig=plt.figure()
range=np.arange(10,140,10)
ax=fig.add_axes([0,0,1,1])
ax.scatter(range,girls,label='girls',color='r',s=100,marker='*')#s=marker size
ax.scatter(range,boys,color='b',edgecolors='g')
ax.set_xlabel('range')
ax.set_ylabel('No of students')
plt.legend()
plt.show()

```

```
#twin axes
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])

x=np.arange(1,11)
ax1.plot(x,np.log10(x),color='g',marker='s')
ax1.set_ylabel('log')
ax1.set_xlim(0,20)


ax2=ax1.twinx()
ax2.plot(x,np.sqrt(x),color='r')

ax2.set_ylabel('sqrt')
plt.legend(loc='upper right',labels=('log','sqrt'))
plt.show()
```
