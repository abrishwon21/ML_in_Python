```
import pandas as pd
import numpy as np
import seaborn as sb
df=sb.load_dataset('titanic')
df.head()
```

```
#Today we will be talking about the seaborn
sb.barplot(x='sex',y="survived",hue="class",data=df)
```

```
sb.countplot(x='class',data=df,palette="Blues")
#countplot=histogram
```

```
#now let's pointplot
sb.pointplot(x="sex",y="survived",hue="class",palette='RdYlBu',data=df)
```
